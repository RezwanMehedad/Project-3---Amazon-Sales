```python
#Finding Size
ax=sns.countplot(x = 'Size', data = df)
```

```python
ax=sns.countplot(x = 'Size', data = df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#most of the people are buying Size Medium
```

```python
df.groupby(['Size'], as_index = False)['Qty'].sum().sort_values(by = 'Qty', ascending = False)
```

```python
S_Qty = df.groupby(['Size'], as_index = False)['Qty'].sum().sort_values(by = 'Qty', ascending = False)

sns.barplot(x = 'Size', y = 'Qty', data = S_Qty)
```

```python
# Most Quantity was sold on Medium Size
```

```python
sns.countplot(data = df, x='Courier Status', hue = 'Status')
```

```python
plt.figure(figsize = (10,5))

ax=sns.countplot(data=df, x = "Courier Status", hue = 'Status')

plt.show()
```

```python
df['Size'].hist()
```

```python
df['Category'] = df['Category'].astype(str)
column_data = df['Category']
plt.figure(figsize = (10,5))

plt.hist(column_data, bins = 30, edgecolor = 'Black')
plt.xticks(rotation = 45)
plt.show()
```

```python
# Tshirt has been sold the most but shirt is almost equally preferred to customers
```

```python
B2B_Check = df['B2B'].value_counts()

plt.pie(B2B_Check, labels = B2B_Check.index, autopct = '%1.1f%%')

plt.show()
```

```python
# The maximum number of buyers are retailers and 0.8% are B2B buyers
```

```python
x_data = df['Category']
y_data = df['Size']

plt.scatter(x_data, y_data)
plt.xlabel('Category')
plt.ylabel('Size')
plt.title('Products to Sizes')

plt.show()
```

```python
plt.figure(figsize = (15,8))

sns.countplot(data=df, x = 'ship-state')

plt.xlabel('ship-state')
plt.ylabel('count')
plt.title('Distribution to State')
plt.xticks(rotation = 90)
plt.show()
```

```python
top_10_state = df['ship-state'].value_counts().head(10)

plt.figure(figsize = (15,8))

sns.countplot(data=df[df['ship-state'].isin(top_10_state.index)], x = 'ship-state')

plt.xlabel('ship-state')
plt.ylabel('count')
plt.title('Distribution to State')
plt.xticks(rotation = 45)
plt.show()
```

```python
# Majority of the buyers are from Maharashtra
```

Conclusion: The Data Analysis reveals that the business has a significant customer base in Maharashtra state, mainly serves retailers, fulfills orders through Amazon, experiences high demand in Tshirt and seees M-Size as the preferred choice among buyers
